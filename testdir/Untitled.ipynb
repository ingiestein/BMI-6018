```python
import pandas as pd
import numpy as np
import datetime

df = pd.DataFrame(
    {
        "A": ["one", "one", "two", "three"] * 6,
        "B": ["A", "B", "C"] * 8,
        "C": ["foo", "foo", "foo", "bar", "bar", "bar"] * 4,
        "D": np.random.randn(24),
        "E": np.random.randn(24),
        "F": [datetime.datetime(2013, i, 1) for i in range(1, 13)]
        + [datetime.datetime(2013, i, 15) for i in range(1, 13)],
    }
)
df
```

```python
f = pd.pivot_table(df,index=['A','B'], columns="C", values="D")
f
```

```python
pf = pd.pivot_table(df,index=['A','B'], columns="C", values="D",aggfunc=['sum', 'mean', 'count']).copy()
pf
```
