```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.cluster import KMeans
from sklearn.compose import ColumnTransformer
import time
```

```python
diabetes_data = pd.read_csv('dmdata/diabetic_data.csv')

#remove identifer columns
diabetes_data = diabetes_data.drop(['encounter_id','patient_nbr','payer_code','readmitted'],axis=1, errors='ignore')

(n_samples, n_features) = diabetes_data.shape
print(n_samples,n_features)

#find numeric/categorical columns

numeric_cols = diabetes_data.select_dtypes(include=['int64','float64']).columns.tolist()
categorical_cols = diabetes_data.select_dtypes(include=['object']).columns.tolist()
```

```python
# now convert categorical columns to numerical ones using "OneHotEncoder" method
# normalize numerical columns be the same size.

preprocesser = ColumnTransformer([
                ('num', StandardScaler(), numeric_cols),
                ('cat', OneHotEncoder(drop='first',sparse_output=False),categorical_cols)
    
])

#apply to df
processed_data = preprocesser.fit_transform(diabetes_data)
```

```python

print(processed_data.shape)
```

```python
#Figure out elbow
print("Starting elbow calculation.")
k_len = range(1,15)
inertia = []
start = time.perf_counter()
for k in k_len:
    k_means = KMeans(n_clusters=k)
    k_means.fit(processed_data)
    inertia.append(k_means.inertia_)
# %% raw
end = time.perf_counter()
print(f"Elapsed: {end - start:.6f} seconds")
print("inertia",inertia)
```
